```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
```

```python
df_chinese = pd.read_csv("chinese_frequencies.csv")
df_english = pd.read_csv("english_frequencies.csv")
```

```python
y=df_chinese['ln_frequency']
x=df_chinese['ln_rank']
y1=df_english['ln_frequency']
x1=df_english['ln_rank']
```

```python
x_line = np.linspace(0,10,200)
y_line=10-x_line
```

```python
fig, ax = plt.subplots()
ax.plot(x,y,label = 'Chinese')
ax.plot(x1,y1, label = 'English')
ax.plot(x_line,y_line,label='Slope = -1')
ax.legend()
plt.show()
```
